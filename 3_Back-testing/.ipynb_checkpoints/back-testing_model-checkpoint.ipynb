# Data type manipulate

Python also includes a data type for sets. 
A set is an unordered collection with no duplicate elements. 
Basic uses include membership testing and eliminating duplicate entries. 
Set objects also support mathematical operations like union, intersection, difference, and symmetric difference.

# create the rebalance matrix:

```python
#get weight table from mean variance
#weight = pd.read_excel('weights_table_mean_variance.xlsx')
weight = pd.read_excel('mean_weighted_user8.xlsx')

#get unique stock name from table
stocks_name = list(set(weight['tic']))
stocks_name.sort()
```

```python
#get unique trade date from table
tradedate = list(set(weight['trade_date']))
tradedate.sort()
```

```python
#add the last trade date 20170901
tradedate_full = tradedate.copy()
tradedate_full.append(20170901)
```

```python
#construct an empty matrix
#rows are stocks: 971 unique stocks
#columns are trade dates
#name the index as stock names
balance_weight = pd.DataFrame(index = stocks_name, columns = tradedate)
balance_weight = balance_weight.fillna(0)
```

```python
balance_weight.head()
```

```python
weight.head()
```

```python
range(weight.shape[0])
```

```python
#loop through the weights table to file the matrix
for i in range(weight.shape[0]):
    tics = weight.loc[i, 'tic']
    date = weight.loc[i, 'trade_date']
    w = weight.loc[i, 'weights']
    balance_weight.loc[tics, date] = w
    
```

```python
# weighted
# mean-variance shares matrix
balance_share = pd.DataFrame(index = stocks_name, columns = tradedate)
balance_share['initial'] = 0
balance_share = balance_share.fillna(0)
#mean-variance cost matrix
balance_cost = pd.DataFrame(index = stocks_name, columns = tradedate)
balance_cost = balance_cost.fillna(0)
#mean-variance cash matrix
balance_cash = pd.DataFrame(index = stocks_name, columns = tradedate)
balance_cash = balance_cash.fillna(0)

#mean-variance portfolio value
portfolio = pd.DataFrame(index = range(1), columns = tradedate_full)
portfolio.loc[0,tradedate[0]] = capital
```

```python
portfolio
```

```python
# equally weight
weight_tmp = weight.copy()
for i in tradedate:
    ind = np.where(weight_tmp['trade_date'] == i)[0]
    we = 1/len(ind)
    weight_tmp.loc[ind, 'weights'] = we

# create the equally weighted matrix:
equal_weight = pd.DataFrame(index = stocks_name, columns = tradedate)
equal_weight = equal_weight.fillna(0)

for i in range(weight_tmp.shape[0]):
    tics = weight_tmp.loc[i, 'tic']
    date = weight_tmp.loc[i, 'trade_date']
    w = weight_tmp.loc[i, 'weights']
    equal_weight.loc[tics, date] = w
import xlsxwriter
#writer2 = pd.ExcelWriter('equal_weight.xlsx')
writer2 = pd.ExcelWriter('equal_weight_user8.xlsx')

equal_weight.to_excel(writer2,'Sheet1')
writer2.save()
```

```python
# the equally weight
balance_share_equal = pd.DataFrame(index = stocks_name, columns = tradedate)
balance_share_equal['initial'] = 0
balance_share_equal = balance_share_equal.fillna(0)
balance_cost_equal = pd.DataFrame(index = stocks_name, columns = tradedate)
balance_cost_equal = balance_cost_equal.fillna(0)
balance_cash_equal = pd.DataFrame(index = stocks_name, columns = tradedate)
balance_cash_equal = balance_cash_equal.fillna(0)

portfolio_equal = pd.DataFrame(index = range(1), columns = tradedate_full)
portfolio_equal.loc[0,tradedate[0]] = capital
```
